```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ===== CSV =====
df_csv = pd.read_csv("world_happiness_report.csv")

# 国名・スコアがある行だけ
df_csv = df_csv.dropna(subset=["Country", "Happiness Score"])

# スコア順に並べて順位付け
df_csv = df_csv.sort_values(by="Happiness Score", ascending=False)
df_csv["CSV_Rank"] = range(1, len(df_csv) + 1)

# 英語 → 日本語
country_en_to_ja = {
    "Australia": "オーストラリア",
    "Canada": "カナダ",
    "Denmark": "デンマーク",
    "Finland": "フィンランド",
    "Iceland": "アイスランド",
    "Netherlands": "オランダ",
    "New Zealand": "ニュージーランド",
    "Norway": "ノルウェー",
    "Sweden": "スウェーデン",
    "Switzerland": "スイス"
}

df_csv["Country_JA"] = df_csv["Country"].apply(
    lambda x: country_en_to_ja.get(x, x)
)

# ===== Excel =====
df_xls = pd.read_excel("country_scores.xlsx")

# ===== 共通国 =====
common = set(df_csv["Country_JA"]) & set(df_xls["Country"])

csv_common = df_csv[df_csv["Country_JA"].isin(common)][
    ["Country_JA", "CSV_Rank"]
]

xls_common = df_xls[df_xls["Country"].isin(common)][
    ["Country", "Rank"]
]

# 結合
merged = pd.merge(
    csv_common,
    xls_common,
    left_on="Country_JA",
    right_on="Country"
).sort_values("CSV_Rank")

# ===== グラフ =====
plt.figure()

plt.scatter(merged["CSV_Rank"], merged["Rank"])

for i, (_, row) in enumerate(merged.iterrows(), start=1):
    plt.text(row["CSV_Rank"], row["Rank"], str(i), fontsize=9)

plt.xlabel("幸福度の順位")
plt.ylabel("健康寿命の順位")
plt.title("共通国の順位比較（番号対応）")
plt.grid(True)
plt.tight_layout()
plt.show()

# 対応表
display(merged[["Country_JA"]].reset_index(drop=True).assign(No=lambda x: x.index+1))

```
