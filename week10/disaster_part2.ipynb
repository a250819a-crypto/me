```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Gothic'

# CSVファイルからデータセットを読み込む
csv_file_path = 'emdat.csv'  # CSVファイルのパスに置き換えてください
emdat = pd.read_csv(csv_file_path)

# データセットの最初の数行を表示
emdat.head()
```

```python
def plot_event_vs_deaths(country, top_n=5):
    """
    指定した国について、
    災害タイプ別の発生件数と死亡者数を同時に可視化する関数
    """

    # 国でデータ抽出
    country_data = emdat[emdat["Country"] == country]

    # 災害タイプ別に集計
    summary = country_data.groupby("Disaster Type").agg(
        発生件数=("Disaster Type", "count"),
        死亡者数=("Total Deaths", "sum")
    )

    # 発生件数が多い順に上位N件
    summary = summary.sort_values("発生件数", ascending=False).head(top_n)

    # グラフ描画
    summary.plot(kind="bar", figsize=(10, 5))
    plt.title(f"{country}：災害タイプ別 発生件数と死亡者数")
    plt.xlabel("災害タイプ")
    plt.ylabel("件数 / 死亡者数")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

```python
plot_event_vs_deaths("Japan", top_n=5)


```

```python
plot_event_vs_deaths("United States of America", top_n=5)
```

```python
countries = [
    "Japan",
    "United States of America",
    "India"
]

for country in countries:
    plot_event_vs_deaths(country, top_n=5)
```

日本で起こった災害の中でどの災害の死者数が多いのかをまとめました。
地震の多い日本では地震による死者数が多く、アメリカでは嵐による死者数が多く、インドでは流行り病の死者数が多い。
日本は地震が多く、アメリカでは規模の大きい嵐が多く、インドの流行り病の死者数が多いのは衛生環境が不安定だから死者数がと思った。
