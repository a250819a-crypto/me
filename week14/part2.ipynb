```python
import pandas as pd

# CSVファイルの読み込み
df = pd.read_csv("world_happiness_report.csv")

# 列名を確認（最初に一度実行すると安全）
print(df.columns)

# 仮定：
# 国名の列 → "Country"
# 幸福度の列 → "Happiness Score"
# ※ 実際の列名が違ったらここを書き換える

top20 = (
    df.sort_values(by="Happiness Score", ascending=False)
      .head(20)
      [["Country", "Happiness Score"]]
)

print(top20)
```

```python
import pandas as pd

# Excelファイルを読み込む
df = pd.read_excel("country_scores.xlsx")

# 国名と幸福度がある行だけ残す
df = df.dropna(subset=["Country", "Happiness Score"])

# 幸福度スコアで降順ソートして上位20件
top20 = (
    df.sort_values(by="Happiness Score", ascending=False)
      .head(20)
      .reset_index(drop=True)
)

# 順位を付ける
top20.index += 1

# 結果表示
print(top20[["Country", "Happiness Score"]])
```

```python
import pandas as pd

# CSV読み込み
df = pd.read_csv("world_happiness_report.csv")

# 英語 → 日本語 国名対応表（例）
country_ja = {
    "Finland": "フィンランド",
    "Denmark": "デンマーク",
    "Iceland": "アイスランド",
    "Switzerland": "スイス",
    "Netherlands": "オランダ",
    "Norway": "ノルウェー",
    "Sweden": "スウェーデン",
    "Luxembourg": "ルクセンブルク",
    "New Zealand": "ニュージーランド",
    "Austria": "オーストリア",
    "Australia": "オーストラリア",
    "Canada": "カナダ",
    "Ireland": "アイルランド",
    "Germany": "ドイツ",
    "Belgium": "ベルギー",
    "United States": "アメリカ合衆国",
    "United Kingdom": "イギリス",
    "Czech Republic": "チェコ",
    "France": "フランス",
    "Japan": "日本"
}

# 上位20か国
top20 = df.sort_values(by="Happiness Score", ascending=False).head(20)

# 日本語名列を追加
top20["国名（日本語）"] = top20["Country"].map(country_ja)

print(top20[["国名（日本語）", "Happiness Score"]])
```
