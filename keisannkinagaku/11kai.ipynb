```python
import math

a = 1.0
b = 1e10
c = 1.0

# 公式(1)
x1 = (-b + math.sqrt(b*b - 4*a*c)) / (2*a)
x2 = (-b - math.sqrt(b*b - 4*a*c)) / (2*a)

# 公式(2)
x1_stable = (-b - math.copysign(math.sqrt(b*b - 4*a*c), b)) / (2*a)
x2_stable = c / (a * x1_stable)

print("公式(1):", x1, x2)
print("公式(2):", x1_stable, x2_stable)
```
